```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("data/lab1/data.xlsx")
data.head()
```

```python
plt.figure(figsize=(10, 7), dpi=80)
sns.histplot(data[data["Gender"] == "Male"]["Height"], kde=True)
plt.title("Distributon of male height")
```

```python
plt.figure(figsize=(10, 7), dpi=80)
sns.histplot(data[data["Gender"] == "Female"]["Height"], kde=True)
plt.title("Distributon of female height")
```

```python
plt.figure(figsize=(10, 7), dpi=80)
sns.histplot(data[data["Gender"] == "Female"]["Weight"], kde=True)
plt.title("Distributon of female weight")
```

```python
plt.figure(figsize=(10, 7), dpi=80)
sns.histplot(data[data["Gender"] == "Male"]["Weight"], kde=True)
plt.title("Distributon of male weight")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("./data/lab6/data.xlsx",index_col="№ п/п")
data.head()
```

```python
plt.figure(figsize=(10, 8))
plt.title("Тепловая карта для датасета из лабораторной №6.")
sns.heatmap(
    data.corr(),
    annot=True,
    fmt=".2f",
    vmin=-1,
    vmax=1,
    annot_kws={"size": 7},
)
```
